### 📖 Concept: Detecting Missing Data
Real-world data often has missing values (shown as `NaN`). Use `isna()` and `notna()` to detect them.

```python
# [DEMO] Detecting missing values
data_with_nan = pd.Series([1, np.nan, 3.5, np.nan, 7])

print("Original data:")
display(data_with_nan)

print("\nWhich values are missing?")
display(data_with_nan.isna())

print("\nWhich values are NOT missing?")
display(data_with_nan.notna())
```

```python
print("Strings are considered as object dtype, and they do not have NaN values, so isna() will return False for all elements")
series_with_strings = pd.Series(['test', 'test2', 'test3', ])
print(series_with_strings.isna())
```

## 🎯 Section 2: Indexing and Selection (60 min)

### Learning Objectives (LO2)
By the end of this section, you will be able to:
- **Differentiate** between label-based (`loc`) and position-based (`iloc`) indexing.
- **Apply** boolean filtering to select specific rows.
- **Drop** rows or columns from a DataFrame.
- **Modify** specific data points using selection.

```python
# Setup for Demo
data = pd.DataFrame(np.arange(16).reshape((4, 4)),
                    index=["Ohio", "Colorado", "Utah", "New York"],
                    columns=["one", "two", "three", "four"])
data
```

```python
# [DEMO] Accessing a row by label using .loc
data.loc["Colorado"]
```

### 💡 Concept Demo: Loc vs Iloc
- **.loc[row_label, col_label]**: "Address" based.
- **.iloc[row_pos, col_pos]**: "GPS" based (integers).

```python
# [DEMO] Select 'Colorado' row using loc
print("--- Row by Label ---")
display(data.loc["Colorado"])

# [DEMO] Select 'Colorado' row using iloc (it's at index 1)
print("\n--- Row by Position ---")
display(data.iloc[1])
```

```python
# [DEMO] Boolean Filtering
# Select rows where column 'three' is greater than 5
data[data["three"] > 5]
```

### 📖 Concept: Dropping Rows and Columns
Use the `drop()` method to remove rows or columns. By default, it drops rows. Use `axis=1` or `axis="columns"` to drop columns.

```python
# [DEMO] Dropping rows and columns
print("Drop rows 'Colorado' and 'Ohio':")
display(data.drop(["Colorado", "Ohio"]))

print("\nDrop column 'two':")
display(data.drop("two", axis=1))

# These operations do not affect the original DataFrame unless we specify 'inplace=True:'
```

## 🎯 Section 3: Function Application, Sorting, & Ranking (60 min)

### Learning Objectives (LO3 & LO4)
By the end of this section, you will be able to:
- **Sort** data by index or values.
- **Rank** data to handle ties.
- **Apply** custom functions to transform data.

```python
# Setup Data
frame = pd.DataFrame({
    "b": [4, 7, -3, 2], 
    "a": [0, 1, 0, 1]
}, index=['Row1', 'Row2', 'Row3', 'Row4'])
frame
```

### 💡 Concept Demo: Sorting & Ranking
1. `sort_index()`
2. `sort_values(by=...)`
3. `rank()`

```python
# [DEMO] Sorting by index
sort_index = frame.sort_index()
display(sort_index)
```

```python
# [DEMO] Sort by values in column 'b'
frame.sort_values(by="b")
```

```python
# [DEMO] Sort by values in column 'b' in descending order
frame.sort_values(by="b", ascending=False)
```

```python
print("Original Frame:")
display(frame)
# rank by column 'b' using 'min' method for ties and descending order
print("Ranked Frame:")
frame.rank(method='min', ascending=False)
```

```python
# [DEMO] Applying a function so that if value in column 'b' is greater than 2, label it as "Big", otherwise "Small"

def label_size(val):
    return "Big" if val > 2 else "Small"

frame['b_size'] = frame['b'].apply(label_size)
frame
```

```python
# applying the same function on rows, use apply with axis=1 to apply the function to each row instead of each column

def label_size(row):
    if row['b'] > 2:
        return "Big"
    else:
        return "Small"
frame['b_size_row'] = frame.apply(label_size, axis=1)
frame
```

### 🛠️ Activity: Leaderboard Logic

**Scenario:** You have exam scores. You need to rank them and assign a Pass/Fail status.

**Tasks:**
1. Run the starter code to generate the `scores` DataFrame.
2. Sort the DataFrame by 'Math' score in descending order (highest first).
3. Create a new column 'Math_Rank' using the `.rank()` method.
4. Define a function that returns "Pass" if score >= 60 and "Fail" otherwise.
5. Apply this function to the 'Math' column to create a new 'Status' column.

```python
# [EXERCISE] Starter Code

scores = pd.DataFrame({
    "Student": ["Alice", "Bob", "Charlie", "David"],
    "Math": [85, 45, 92, 60],
    "Science": [70, 55, 88, 72]
})
print("Original Scores DataFrame:")
display(scores)

# Your code here

# 2. Sort the DataFrame by 'Math' score in descending order (highest first).
print("Sorted DataFrame by 'Math' score in descending order (highest first):")
sort_index = scores.sort_values(by="Math", ascending=False)
display(sort_index)

# 3. Create a new column 'Math_Rank' using the `.rank()` method in descending order so that the highest score gets rank 1, as integer ranks, where the new 'Math_Rank' column sorted by 'Math_Rank' in ascending order (rank 1 at the top).
print("DataFrame with 'Math_Rank' column:")
scores['Math_Rank'] = scores['Math'].rank(ascending=False).astype(int)
scores = scores.sort_values(by="Math_Rank")
display(scores)

# 4. Define a function that returns "Pass" if score >= 60 and "Fail" otherwise.
def evaluate_status(score):
    if score >= 60:
        return "Pass"
    else:
        return "Fail"

# 5. Apply this function to the 'Math' column to create a new 'pass or fail' column.
print("DataFrame with 'pass or fail' column:")
scores['Status'] = scores['Math'].apply(evaluate_status)
display(scores)
```
