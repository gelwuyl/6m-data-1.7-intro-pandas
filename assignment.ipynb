```python
import pandas as pd
import numpy as np

books = pd.DataFrame({
    'title': ['The Pragmatic Programmer', 'Clean Code', 'Python Crash Course',
              'Data Science from Scratch', 'Designing Data-Intensive Apps'],
    'author': ['Hunt & Thomas', 'Robert Martin', 'Eric Matthes',
               'Joel Grus', 'Martin Kleppmann'],
    'genre': ['Tech', 'Tech', 'Tech', 'Data', 'Data'],
    'price': [45.99, 38.50, 29.99, 42.00, 55.00],
    'units_sold': [120, 95, 200, 80, 60],
    'in_stock': [True, True, False, True, False]
})
```

## Challenge 1: Build and Inspect the Dataset

Tasks:

    Use .info() — how many non-null values does each column have?
    Use .describe() — what's the average price and maximum units sold?
    Add a revenue column: price × units_sold.
    What is the total revenue across all books?

```python
print("Books DataFrame:")
display(books)
```

```python
# how many non-null values in each column. 

non_null_counts = books.notnull().sum().to_frame().T.rename(index={0: 'non_null_count'})
# 'to_frame' displays the result as a table with column names and non-null counts
# 'T transposes the table to show column names as headers and non-null counts as a single row.
# 'rename' changes the index name from 0 to 'non_null_count' for better readability.

display(non_null_counts)
```

```python
# what is the average price. what is maximum units sold.
display(books.describe())

average_price_of_books = books['price'].mean()
maximum_units_sold = books['units_sold'].max()
print(f"Average Price of Books: ${average_price_of_books:.2f}")
print(f"Maximum Units Sold: {maximum_units_sold}")
```

```python
# revenue is price * units_sold and add it as a new column 'revenue'

books['revenue'] = books['price'] * books['units_sold']
print("Books DataFrame with 'revenue' column:")
display(books)
```

```python
# total revenue for all the books

total_revenue = books['revenue'].sum()
print(f"Total Revenue for All Books: ${total_revenue:.2f}")
```

## Challenge 2: Indexing & Filtering

Tasks:

    Use `.loc` to select only `title` and `revenue` for Data genre books.
    Use `.iloc` to retrieve the last 3 rows and first 3 columns.
    Filter books that are **in stock** AND have a **price above £40**.
    Find the book with the **highest revenue** using `.loc` with `.idxmax()`.

```python
1.
# Use .loc to show only title and revenue for Data genre books.
print("Title and Revenue for Data Genre Books:")
display(books.loc[books['genre'] == 'Data', ['title', 'revenue']])


2.
# Use .iloc to retrieve the last 3 rows and first 3 columns.
print("Last 3 Rows and First 3 Columns:")
display(books.iloc[-3:, :3])


3.
# Filter books that are in stock AND have a price above £40.
filtered_books = books[(books['in_stock'] == True) & (books['price'] > 40)]
print("Books that are in stock AND have a price above £40:")
display(filtered_books)


4.
# Find the book with the highest revenue using .loc with .idxmax(). 
highest_revenue_book = books.loc[[books['revenue'].idxmax()]] 
# use double brackets [[...]] which return a DataFrame (not a Series), so it displays with the column names as headers and the data as a single row—exactly like the original books table format.
print("Book with the Highest Revenue:")
display(highest_revenue_book)
```

## Challenge 3: Transform, Sort & Rank

```python
# 1.    Use .apply() to add a price_tier column: 'Budget' (< £35), 'Standard' (£35–£50), 'Premium' (> £50).
# define a function to categorize price
def categorize_price(price):
    return 'premium' if price > 50 else 'standard' if 35 <= price <= 50 else 'budget'

# apply the function to the 'price' column and create a new 'price_tier' column
books['price_tier'] = books['price'].apply(categorize_price)
print("Books DataFrame with 'price_tier' column:")
display(books)


# 2.    Sort the DataFrame by revenue descending.
# sort the books by revenue in descending order
sorted_books = books.sort_values(by='revenue', ascending=False)
print("Books sorted by revenue (descending):")
display(sorted_books)


# 3.    Add a sales_rank column using .rank(ascending=False).
# add a new column 'sales_rank' that ranks the books based on revenue, with 1 being the highest revenue.
books['sales_rank'] = books['revenue'].rank(ascending=False, method='min').astype(int)

# sort sales_rank in ascending order to show the highest revenue book at the top
print("Books with 'sales_rank' column, sorted by sales rank:")
display(books.sort_values(by='sales_rank')) 


# 4.    Use .map() to encode genre as a number: {'Tech': 0, 'Data': 1}.
# create a mapping dictionary for genre encoding
genre_mapping = {'Tech': 0, 'Data': 1}
books['genre'] = books['genre'].map(genre_mapping)
print("Books DataFrame with 'genre_encoded' column:")
display(books)
```

## 🏆 Stretch Challenge (Optional)

The manager wants a report comparing average price and total revenue by price_tier. Use groupby() to generate this summary. Which tier contributes the most total revenue despite having the lowest average price?

```python
summary = books.groupby('price_tier').agg(
    average_revenue=('revenue', 'mean'),
    total_revenue=('revenue', 'sum')
)
display(summary)
```
